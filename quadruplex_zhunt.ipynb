## Ищем квадруплексы и сохраняем в bed формат

```python
input_file ='/content/715/GCF_000190715.1_v1.0/GCF_000190715.1_v1.0_genomic.fna'
fasta_sequence = SeqIO.parse(input_file,'fasta')
PQS = []
PQS_minus = []
pattern="(G{3,5}[ATGC]{1,7}){3,}G{3,5}" #шаблон стандартного квадруплекса на + стренде
pattern_minus="(C{3,5}[ATGC]{1,7}){3,}C{3,5}" #на - стренде
for record in SeqIO.parse(input_file, "fasta"):
    name, sequence = record.id, str(record.seq)
    PQS += [[name, m.start(),m.end(),m.group(0), name, '+'] for m in re.finditer(pattern,sequence,re.IGNORECASE)] #игнорируем регистр
    PQS_minus += [[name, m.start(),m.end(),m.group(0), name, '-'] for m in re.finditer(pattern_minus,sequence,re.IGNORECASE)]
print(len(PQS))
print(len(PQS_minus))
```

```python
quadr = pd.DataFrame(PQS+PQS_minus, columns=['Chromosome', 'Start', 'End', 'Sequence', 'Name', 'Strand']) ##тут важно чтобы на первой позиции было название скаффолда, на 2 и 3 координаты, а на 6 стренд
quadr
```

```python
bf = pybed.BedFrame.from_frame(meta=[], data=quadr)
bf.to_file('quadr.bed')
```

## Переведем аннотацию в необходимый формат и вытащим bed файлы промоторов/экзонов/межгенников/интронов/downstream

```python
def calculate_scaffold_lengths(fasta_file, output_file):
    with open(output_file, 'w') as out:
        for record in SeqIO.parse(fasta_file, "fasta"):
            scaffold_name = record.id
            scaffold_length = len(record.seq)
            out.write(f"{scaffold_name}\t{scaffold_length}\n")
calculate_scaffold_lengths('/content/715/GCF_000190715.1_v1.0/GCF_000190715.1_v1.0_genomic.fna', 'scaffolds.sizes')
```

## Квадруплексы

```python
def count_unique(file_path):
    unique_lines = set()  # Используем множество для хранения уникальных строк
    try:
        with open(file_path, 'r') as file:
            for line in file:
                unique_lines.add(line.strip())  # Удаляем пробелы и добавляем строку в множество
    except FileNotFoundError:
        print(f"Файл {file_path} не найден.")
        return 0
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return 0

    return len(unique_lines)  # Возвращаем количество уникальных строк
```

```python
from Bio import SeqIO

input_fasta = '/content/715/GCF_000190715.1_v1.0/GCF_000190715.1_v1.0_genomic.fna'
output_dir = '/content/drive/MyDrive/Bioinf_files/scaffolds/'

def split_fasta_by_scaffolds(input_fasta, output_dir):
    fasta_sequences = SeqIO.parse(open(input_fasta),'fasta')
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        with open(output_dir + name + ".fasta", "w") as output_file:
            output_file.write(f">{name}\n{sequence}\n")

split_fasta_by_scaffolds(input_fasta, output_dir)
```

```python
import os
directory = "/content/drive/MyDrive/Bioinf_files/scaffolds"
results = []
for filename in os.listdir(directory):
    if filename.endswith(".fasta.Z-SCORE"):
        clean_filename = filename.rstrip('.fasta.Z-SCORE')
        filepath = os.path.join(directory, filename)
        df = pd.read_table(filepath, delim_whitespace=True, skiprows=1,
                           names=['Start', 'End', 'None1', 'None2', 'None3', 'Z-score', 'Sequence', 'None4'])
        filtered_df = df[df['Z-score'] >= 400]
        for index, row in filtered_df.iterrows():
            results.append((clean_filename, row['Start'], row['End']))
```

```python
zdna = pd.DataFrame(results, columns=['Chromosome', 'Start', 'End'])
zdna
```
